```python
import  numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
df.head()
```

<h2> Multi-variable Linear Regression Using Gradient Descent

```python
class LinearRegression_GD:
  def __init__(self , learning_rate = 0.01 , epochs = 1000):
    self.learning_rate = learning_rate
    self.epochs = epochs
    self.m = None
    self.b = None
  def fit(self , X_train , y_train):
      # Set a random value to m and b. Let's say 0
      self.m , self.b = 0 , 0
      n = len(X_train)

      # Do the iterations
      for epoch in range(self.epochs):
         # predict y using current m and b
         y_pred = self.m * X_train + self.b
         # Find the error
         error = y_train - y_pred

         # Find slope with respect to m and b
         dm = (-2/n) * np.sum(error * X_train)
         db = (-2/n) * np.sum(error)

         # Update m and b
         self.m = self.m - self.learning_rate * dm
         self.b = self.b - self.learning_rate * db
  # make the predict method
  def predict(self , X_test):
    return self.m * X_test + self.b
```

<h2>Make The Train and Test Set

```python
X = df.iloc[: , 0].values
y = df.iloc[: , 1].values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size = 0.2 , random_state = 2)
```

```python
my_lr = LinearRegression_GD()
```

```python
my_lr.fit(X_train , y_train)
```

```python
y_pred_my_lr = my_lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test , y_pred_my_lr)
```

```python
m = my_lr.m
b = my_lr.b

print(f'The value of m is {m}')
print(f'The value of b is {b}')
```

<h2> Let's Check the performs with sklearn model

```python
from sklearn.linear_model import LinearRegression
sk_lr = LinearRegression()
```

```python
sk_lr.fit(X_train.reshape(-1 , 1) , y_train)
```

```python
y_pred_sklearn = sk_lr.predict(X_test.reshape(-1 , 1))
```

```python
print(f'The value of m is {sk_lr.coef_}')
print(f'The value of b is {sk_lr.intercept_}')
```

```python
print(f"The r2 score of my model is {r2_score(y_test , y_pred_my_lr)}")
print(f"The r2 score of sklearn model is {r2_score(y_test , y_pred_sklearn)}")
```
