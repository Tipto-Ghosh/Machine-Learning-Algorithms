```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

np.set_printoptions(precision = 4 , suppress = True)
```

```python
# Let's make a polynomial Dataset
np.random.seed(42)
# make a single col X
X_values = np.linspace(0, 4 * np.pi, 200)  # X ranges over multiple wave cycles
y_values = np.sin(X_values) + np.random.normal(0, 0.2, 200)  # sine wave + noise
```

```python
df = pd.DataFrame({'input_feature' : X_values , 'target' : y_values})
df.head()
```

```python
# Let's see the relation of input_feature and target 
plt.scatter(df['input_feature'] , df['target']) 
# plt.plot(df['input_feature'] , df['target'] , color = 'r')
plt.show()
```

## Split train and test

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split(
        df.drop(columns = ['target']), 
        df['target'], 
        test_size = 0.2,
        random_state = 2
)
```

```python
plt.scatter(X_train , y_train)
plt.show()
```

## Fitting to Linear Model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train , y_train)
```

```python
from sklearn.metrics import r2_score , mean_squared_error
y_lr_pred = lr.predict(X_test) 

print(f"Linear models mse: {mean_squared_error(y_test , y_lr_pred)}")
print(f"Linear models r2_score: {r2_score(y_test , y_lr_pred)}")
```

```python
plt.scatter(df['input_feature'] , df['target'])
plt.plot(X_test , y_lr_pred , color = 'red')
plt.title('Linear models best fit line')
plt.show()
```
